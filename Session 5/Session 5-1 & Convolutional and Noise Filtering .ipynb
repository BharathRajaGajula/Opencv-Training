# Convolutional Filtering and Noise Filtering

```python
# Importing Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the image
img = cv2.imread("musk.jpg")

# Defining the kernel size
kernel_size = 3

for i in range(10):
    kernel = np.random.rand(kernel_size, kernel_size)
    #print(kernel)
    result = cv2.filter2D(img, -1, kernel, (-1, -1), delta=0, borderType=cv2.BORDER_DEFAULT)

plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(img[...,::-1]);plt.title("Original Image")
plt.subplot(122);plt.imshow(result[...,::-1]);plt.title("Convolution Result")
```

# Noise Filtering

```python
# Load Images
saltPepper = cv2.imread('SaltPepperNoise.png')
gaussian = cv2.imread('gaussian_noise.png')

# Convert to rgb
saltPepper = cv2.cvtColor(saltPepper, cv2.COLOR_BGR2RGB)
gaussian = cv2.cvtColor(gaussian, cv2.COLOR_BGR2RGB)
```

```python
# Salt and Pepper noise filtering
saltPepperMedianFiltering = cv2.medianBlur(saltPepper,3)
saltPepperBilateralFiltering = cv2.bilateralFilter(saltPepper,d=10, sigmaColor=200, sigmaSpace=80)
#https://homepages.inf.ed.ac.uk/rbf/CVonline/LOCAL_COPIES/MANDUCHI1/Bilateral_Filtering.html#The%20Idea
#A bilateral filter is a non-linear, edge-preserving, and noise-reducing smoothing filter for images. It replaces the intensity of each pixel with a weighted average of intensity values from nearby pixels. This weight can be based on a Gaussian distribution.

plt.figure(figsize=[20,20])
plt.subplot(131);plt.imshow(saltPepper);plt.title('Salt n Pepper Noise')
plt.subplot(132);plt.imshow(saltPepperMedianFiltering);plt.title('Median Filtering')
plt.subplot(133);plt.imshow(saltPepperBilateralFiltering);plt.title('Bilateral Filtering')
```

```python
# Gaussian Noise filtering
gaussianMedianFiltering = cv2.medianBlur(gaussian, 7)
gaussianBilateralFiltering = cv2.bilateralFilter(gaussian, d=15, sigmaColor=70, sigmaSpace=40)

plt.figure(figsize=[20,20])
plt.subplot(131);plt.imshow(gaussian);plt.title('Gaussian Noise')
plt.subplot(132);plt.imshow(gaussianMedianFiltering);plt.title('Median Filtering')
plt.subplot(133);plt.imshow(gaussianBilateralFiltering);plt.title('Bilateral Filtering')
```

```python
# Gaussian Noise excessive filtering
gHeavyMedianFiltering = cv2.medianBlur(gaussian, 13)
gHeavyBilateralFiltering = cv2.bilateralFilter(gaussian, d=20, sigmaColor=100, sigmaSpace=40)

plt.figure(figsize=[20,20])
plt.subplot(131);plt.imshow(gaussian);plt.title('Gaussian Noise')
plt.subplot(132);plt.imshow(gHeavyMedianFiltering);plt.title('Heavy Median Filtering')
plt.subplot(133);plt.imshow(gHeavyBilateralFiltering);plt.title('Heavy Bilateral Filtering')
```
