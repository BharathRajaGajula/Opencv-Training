```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
elon = plt.imread('musk.jpg')
```

```python
elon.shape
```

```python
filter_image = np.ones((180, 300, 3), dtype = 'uint8')
```

```python
plt.imshow(filter_image)
```

```python
plt.imshow(filter_image*255)
```

```python
filter_image[:,:,0] = filter_image[:,:,0]*255
filter_image[:,:,1] = filter_image[:,:,1]*216
filter_image[:,:,2] = filter_image[:,:,2]*160
```

```python
plt.imshow(filter_image)
```

```python
filtered_elon = cv2.add(elon, filter_image)
```

```python
plt.imshow(filtered_elon)
```
